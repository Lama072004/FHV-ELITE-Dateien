```python
import numpy as np
import matplotlib.pyplot as plt

X = np.genfromtxt("test.csv")
Year = X[1:len(X),0]
x = X[1:len(X),1]
y = X[1:len(X),2]

```

```python
n = len(Year)
print(n)
A = np.column_stack(((np.ones((n,1))),Year))
regx = np.linalg.lstsq(A,x,rcond=None)[0]
print(regx)
regy = np.linalg.lstsq(A,y,rcond=None)[0]
print(regy)
```

```python
plt.plot(Year,x, 'go')
plt.plot(Year,y,'rx')
num = np.linspace(1980,1987,2)
plt.plot(num,num*regx[1]+regx[0],'g')
plt.plot(num,num*regy[1]+regy[0],'r')
```
