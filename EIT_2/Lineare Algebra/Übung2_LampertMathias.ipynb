## 1. Moddellierungsbeispiele

### Geometrie

```python
import numpy as np

A = np.array([
    [2, -5],  # Koeffizienten der ersten Gleichung: 2x - 5y
    [-1, 7]   # Koeffizienten der zweiten Gleichung: -x + 7y
])

b = np.array([8, 5])  # 8 und 5 sind die rechten Seiten der Gleichungen

# Lösen des linearen Gleichungssystems
solution = np.linalg.solve(A, b)

# Ausgabe der Lösung
x, y = solution
print(f"Der Schnittpunkt liegt bei x = {x} und y = {y}")
```

```python
import matplotlib.pyplot as plt

def f(x):
    return (2*x-8)/5
    
def g(x):
    return (5+x)/7

x = np.linspace(-5,15,500)
plt.plot(x, f(x), label=r'$f(x) = \frac{2x - 8}{5}$')  # Funktion f(x)
plt.plot(x, g(x), label=r'$g(x) = \frac{5 + x}{7}$')   # Funktion g(x)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
```

### Kinematik

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameterdarstellungen der Geraden
def P1(t):
    return np.array([1 - 3*t, -2 + 4*t])

def P2(s):
    return np.array([-6 + s, 6 - s])

# Wertebereich für die Parameter t und s
t_values = np.linspace(-5, 5, 100)
s_values = np.linspace(-5, 5, 100)

# Berechnung der Punkte auf den Geraden
P1_points = np.array([P1(t) for t in t_values])
P2_points = np.array([P2(s) for s in s_values])

# Plot der Geraden
plt.plot(P1_points[:, 0], P1_points[:, 1], label=r'$P_1(t)$')
plt.plot(P2_points[:, 0], P2_points[:, 1], label=r'$P_2(s)$')

# Plot-Einstellungen
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)  # x-Achse
plt.axvline(0, color='black', linewidth=0.5)  # y-Achse
```

### 3.

```python
def calculate_angle(v, u):
    dot_product = np.dot(v, u)
    norm_v = np.linalg.norm(v)
    norm_u = np.linalg.norm(u)
    cos_theta = dot_product / (norm_v * norm_u)
    theta = np.arccos(cos_theta)
    return np.degrees(theta)  # Winkel in Grad

# Winkel berechnen
angle_a = calculate_angle(v, a)
angle_b = calculate_angle(v, b)
angle_c = calculate_angle(v, c)
angle_d = calculate_angle(v, d)

# Ergebnisse ausgeben
print(f"Winkel zwischen v und a: {angle_a:.2f} Grad")
print(f"Winkel zwischen v und b: {angle_b:.2f} Grad")
print(f"Winkel zwischen v und c: {angle_c:.2f} Grad")
print(f"Winkel zwischen v und d: {angle_d:.2f} Grad")
```
