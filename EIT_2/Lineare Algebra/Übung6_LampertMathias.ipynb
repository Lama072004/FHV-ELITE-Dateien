## 2. Darstellung einer Basis

```python
import numpy as np

# Basis und Vektor definieren
B = np.array([[3, 1], 
              [1, -2]])  # Basisvektoren als Spalten
v = np.array([5, 4])             # Vektor v

# Überprüfung der Basis
det_B = np.linalg.det(B)
print(f"Determinante von B: {det_B}")
if det_B != 0:
    print("B ist eine Basis (Determinante ≠ 0)")

# Lösung des LGS B*c = v
c = np.linalg.solve(B, v)
print(f"Koeffizienten c = {c}")  # Sollte [ 2., -1.] ausgeben

# Überprüfung
v_reconstructed = B @ c
print(f"Rekonstruierter Vektor: {v_reconstructed}")  # Sollte [5, 4] sein
```

## 3 Koordinatentransformation

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegebene Daten
B = np.array([[1, -1], [2, 1]])  # Basis B (Spaltenvektoren)
D = np.array([[1, 3], [0, 1]])   # Basis D (Spaltenvektoren)
v = np.array([4, 5])             # Vektor v

# Darstellung in Basis B (Lösen von B * v_B = v)
v_B = np.linalg.solve(B, v)      # Lösung: [ 3. -1.]

# Darstellung in Basis D (Lösen von D * v_D = v)
v_D = np.linalg.solve(D, v)      # Lösung: [-11.  5.]

print("Darstellung in Basis B (v_B):", v_B)
print("Darstellung in Basis D (v_D):", v_D)

# Daten
B = np.array([[1, -1], 
              [2, 1]])  # Basis B (jede Spalte ein Vektor: b1=[1,2], b2=[-1,1])
D = np.array([[1, 3], 
              [0, 1]])   # Basis D (d1=[1,0], d2=[3,1])
v = np.array([4, 5])             # Vektor v

# Grafik
plt.figure(figsize=(8, 5))

# Vektor v (grün)
plt.quiver(0, 0, v[0], v[1], color='green', angles='xy', scale_units='xy', scale=1, 
           width=0.008, label='Vektor $v=(4,5)$')

# Basis B - getrennte Aufrufe
plt.quiver(0, 0, B[0,0], B[1,0], color='red', angles='xy', scale_units='xy', scale=1, 
           width=0.005, label='$b_1=(1,2)$')  # Erster Basisvektor b1
plt.quiver(0, 0, B[0,1], B[1,1], color='red', angles='xy', scale_units='xy', scale=1, 
           width=0.005, label='$b_2=(-1,1)$') # Zweiter Basisvektor b2

# Basis D - getrennte Aufrufe
plt.quiver(0, 0, D[0,0], D[1,0], color='blue', angles='xy', scale_units='xy', scale=1, 
           width=0.005, label='$d_1=(1,0)$')  # Erster Basisvektor d1
plt.quiver(0, 0, D[0,1], D[1,1], color='blue', angles='xy', scale_units='xy', scale=1, 
           width=0.005, label='$d_2=(3,1)$')  # Zweiter Basisvektor d2

# Einstellungen
plt.xlim(-2, 6)
plt.ylim(-2, 6)
plt.grid(ls=':', alpha=0.5)
plt.legend()
plt.title('Vektor $v$ in verschiedenen Basen', pad=20)
plt.gca().set_aspect('equal')
plt.tight_layout()
plt.show()
```

## 4 Orthogonale Matrizen

```python
# Definition der Matrix A
A = (1/np.sqrt(2)) * np.array([[1, 1], [-1, 1]])

# Berechnungen
A_transpose = A.T
A_inv = np.linalg.inv(A)
ATA = A_transpose @ A
AAT = A @ A_transpose

# Ausgabe
print("A^T:\n", A_transpose)
print("\nA^{-1}:\n", A_inv)
print("\nA^T A:\n", ATA)
print("\nA A^T:\n", AAT)
```

In dem Beispiel ergibt A^TA und AA^T 2.37114e-17. Da diese Werte sehr klein sind kann 0 angenommen werden. Deshalb stimmt das Ergebnis, dass die Berechnungen den Einheitsvektor ergeben.

## 5 Eigenwerte und -vektoren

## 6 Eigenwerte und -vektoren, Invertierbarkeit, Eindeutigkeit der Lösung

```python
import numpy as np

A = np.array([[1, 0, 2], 
              [0, 2, 0], 
              [0, 1, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenwerte von A:", eigenvalues)
print("Eigenvektoren (Spalten):\n", eigenvectors)

# Überprüfe Eigenwerte
if np.all(eigenvalues != 0):
    print("\nA ist invertierbar, da alle Eigenwerte ≠ 0 sind.")
    # Berechne die Inverse zur Demonstration
    A_inv = np.linalg.inv(A)
    print("Inverse von A:\n", A_inv)
else:
    print("\nA ist nicht invertierbar (mind. ein Eigenwert = 0).")

# Alternativ: Determinante prüfen
det_A = np.linalg.det(A)
print("\nDeterminante von A:", det_A)
```
